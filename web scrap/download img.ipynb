```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
df = pd.read_csv('9k_movies.csv')
```

```python
img_df = df[['index','Img' , 'CoverBg']]
```

```python
import requests
from time import sleep

def download_img(idx, url1, retries=10, delay=2):
    for attempt in range(retries):
        try:
            response1 = requests.get(url1, stream=True, timeout=10)

            if response1.status_code == 200:
                # Download image in chunks
                with open(f'image/bg_cover/{idx}.jpg', 'wb') as file:
                    for chunk in response1.iter_content(chunk_size=8192):
                        if chunk:  # Filter out keep-alive chunks
                            file.write(chunk)
                print(f"Image {idx} downloaded successfully!")
                return  # Exit function if successful
            else:
                print(f"Failed to download image {idx}: HTTP Status {response1.status_code}")

        except (requests.exceptions.RequestException, requests.exceptions.ChunkedEncodingError) as e:
            print(f"Attempt {attempt + 1} failed for image {idx}: {e}")
            sleep(delay)  # Wait before retrying

    print(f"Failed to download image {idx} after {retries} attempts")

# Iterate over DataFrame rows
for i in range(img_df.shape[0]):
    download_img(i, img_df.CoverBg[i])
```
