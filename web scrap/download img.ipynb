```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
df = pd.read_csv('9k_movies.csv')
```

```python


img_df = df[['index','Img' , 'CoverBg']]
```

```python
img_df
```

```python

# URL of the image
url = img_df['Img'][1]

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Open a file in binary write mode
    with open('image/img.jpg', 'wb') as file:
        file.write(response.content)
    print("Image downloaded successfully!")
else:
    print("Failed to download image")
```

```python
def download_img( idx,url):
    response = requests.get(url)
    if response.status_code == 200:
        with open(f'image/cover/{idx}.jpg' , 'wb') as file:
            file.write(response.content)
        print("Image downloaded successfully!")
    else:
        print("Failed to download image")
```

```python
for i in range(img_df.shape[0]):
    download_img(i , img_df.Img[i] )
```
